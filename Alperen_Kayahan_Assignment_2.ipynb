```r
library(tidyverse)
library(magrittr)
data(Seatbelts)
library(DT)
```

```r
datatable(Seatbelts)
class(Seatbelts)
```

```r

Seatbelts_2 <- as.data.frame(Seatbelts)
```

```r
# I want to select drivers, PetrolPrice and kms
```

```r
Seatbelts_2 %>% select(drivers,PetrolPrice, kms)
```

```r
#I want to filter DriversKilled less than 100 and rear more than 300
```

```r
Seatbelts_2 %>% filter(DriversKilled <100 & rear>300)
```

```r
#I want to find Total Killed with sum DriversKilled and VanKilled
```

```r
Seatbelts_2 %<>% mutate(Total_Killed = DriversKilled + VanKilled)
Seatbelts_2
```

```r
#I want to group by their law number and find maximum of Drivers Killed.
```

```r
Seatbelts_2 %>% 
group_by(law) %>% 
summarise ( max_drivers_killed = max(DriversKilled, na.rm=T))
```

```r
Seatbelts_2_ed <- Seatbelts_2 %>% group_by(law,rear) %>% summarise( max_killed = max(DriversKilled, na.rm=T))
Seatbelts_2_ed
```

```r
#I want to wide the data acorrding to law and rear. The wide table took their data from max_killed column which I created before.
```

```r
Seatbelts_2_wide <- Seatbelts_2_ed %>% pivot_wider(id_cols = law, names_from = rear, values_from = max_killed)
Seatbelts_2_wide
```

```r
#I want to longer the data according to law and rear. The data took their inputs from maxkilled.
```

```r
Seatbelts_2_wide %>% pivot_longer(cols = -"law", names_to = "rear2", values_to = "max_killed2", values_drop_na = T)
```

```r
table1 <- tibble(DriversKilled=100:120, bb=LETTERS[1:21])
```

```r
#I created a data table which created with letters and driverskilled number then I left join them.
```

```r
Seatbelts_2 %>% left_join(table1, by = "DriversKilled")
```
